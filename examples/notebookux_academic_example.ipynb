# Exemplo academico com notebookux

Este notebook demonstra a biblioteca `notebookux` em um fluxo de relatorio tecnico/academico. A ideia e manter a logica do trabalho em Python normal e usar a lib apenas para a apresentacao visual.

## Preparacao

Se estiver executando fora do ambiente de desenvolvimento, instale a biblioteca a partir da raiz do projeto com:

```powershell
python -m pip install -e ".[examples]"
```

````python
import matplotlib.pyplot as plt
import pandas as pd

from notebookux import UX as CQ

CQ.setup("blueprint", compact=True)
````

````python
CQ.cover(
    title="Relatorio de Exemplo com NotebookUX",
    subtitle="Notebook academico com componentes prontos",
    meta=[
        ("Disciplina", "Inteligencia Artificial"),
        ("Formato", "Notebook"),
        ("Biblioteca", "notebookux"),
    ],
    bullets=[
        "Capa e secoes padronizadas",
        "Metricas, checklist e fluxo visual",
        "Tabelas, graficos e modulo navegavel",
    ],
    people=[
        ("Nome do estudante", "Curso ou turma"),
    ],
)
````

````python
CQ.section(
    "Visao geral do experimento",
    subtitle="Resumo inicial para orientar a leitura do notebook.",
    kicker="Etapa 1",
)

CQ.badges([
    ("Dataset", "blue"),
    ("Classificacao", "green"),
    ("Validacao", "amber"),
    ("Relatorio", "purple"),
])

CQ.key_values(
    {
        "Fonte": "Dados simulados",
        "Objetivo": "Comparar modelos",
        "Saida": "Metricas e graficos",
    },
    title="Metadados do estudo",
)
````

````python
CQ.metrics(
    [
        ("Amostras", "240", "Registros usados na simulacao", "blue"),
        ("Modelos", "3", "Algoritmos comparados", "green"),
        {"label": "Melhor F1", "value": "0.84", "note": "Maior resultado observado", "color": "purple"},
        {"label": "Pendencias", "value": "1", "note": "Discussao final", "color": "amber"},
    ],
    title="Indicadores principais",
)

CQ.checklist(
    [
        ("Dados carregados", True),
        ("Tabela comparativa montada", True),
        {"label": "Discussao dos resultados", "done": False, "note": "Completar apos revisar os graficos."},
    ],
    title="Checklist do notebook",
)
````

````python
CQ.steps(
    [
        ("Carregar dados", "Obter a base e conferir as colunas.", "done"),
        ("Preparar variaveis", "Separar atributos e alvo.", "done"),
        ("Treinar modelos", "Executar algoritmos sob a mesma metodologia.", "active"),
        ("Analisar resultados", "Comparar metricas e registrar conclusoes.", "pending"),
    ],
    title="Fluxo recomendado",
)
````

````python
resultados = pd.DataFrame(
    {
        "modelo": ["Arvore", "KNN", "SVM"],
        "acuracia": [0.82, 0.79, 0.85],
        "f1_medio": [0.80, 0.78, 0.84],
        "tempo_s": [1.4, 2.1, 4.8],
    }
)

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(resultados["modelo"], resultados["f1_medio"], color=[CQ.color("blue"), CQ.color("green"), CQ.color("purple")])
ax.set_ylim(0, 1)
ax.set_title("F1 medio por modelo")
ax.set_ylabel("F1")

CQ.table(resultados, title="Tabela de resultados", subtitle="Valores simulados para demonstracao")
CQ.figure(fig, title="Comparacao visual dos modelos")
````

````python
CQ.module(
    "Resumo navegavel",
    [
        CQ.screen(
            "Conclusao inicial",
            CQ.callout_html(
                "Modelo com melhor desempenho",
                "Neste exemplo simulado, o SVM apresentou o maior F1 medio entre os modelos comparados.",
                variant="success",
            ),
        ),
        CQ.screen("Metricas", CQ.metrics_html([("Melhor F1", "0.84", "SVM", "purple")], title="Resumo numerico")),
        CQ.screen("Tabela", CQ.table_html(resultados, title="Resultados completos")),
        CQ.screen(
            "Comparacao",
            CQ.compare_html(
                "Antes",
                "Resultados espalhados em varias celulas.",
                "Depois",
                "Resumo visual, navegavel e facil de revisar.",
                title="Organizacao do relatorio",
            ),
        ),
    ],
)
````
